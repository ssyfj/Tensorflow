```python
#https://zybuluo.com/hanbingtao/note/541458
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np
```

```python
mnist = input_data.read_data_sets("MNIST_data/",one_hot=True)

#输入的图片是28×28,将图像也设置为序列进行操作。
n_inputs = 28 #输入一行数据，共28列（每次输入28个数据）----输入神经元的个数
max_time = 28 #一共28行（输入28次）
lstm_size = 100 #隐藏单元
n_classes = 10 #分类数
batch_size = 50 #每批次50个样本
n_batch = mnist.train.num_examples // batch_size #一共多少批次

x = tf.placeholder(tf.float32,[None,784])
y = tf.placeholder(tf.float32,[None,10])

#初始化权值
weights = tf.Variable(tf.truncated_normal([lstm_size,n_classes],stddev=0.1))
biases = tf.Variable(tf.constant(0.1,shape=[n_classes]))

#定义RNN网络
def RNN(X,weights,biases):
    inputs = tf.reshape(X,[-1,max_time,n_inputs]) #将图像转换
    #定义LSTM基本CELL  #https://zhuanlan.zhihu.com/p/43041436
    lstm_cell = tf.contrib.rnn.BasicLSTMCell(lstm_size)
    #返回值：元组（outputs, states）
    #outputs：outputs很容易理解，就是每个cell会有一个输出
    #states：states表示最终的状态，也就是序列中最后一个cell输出的状态。
    #一般情况下states的形状为 [batch_size, cell.output_size ]，
    #但当输入的cell为BasicLSTMCell时，state的形状为[2，batch_size, cell.output_size ]，
    #其中2也对应着LSTM中的cell state和hidden state。
    #https://zhuanlan.zhihu.com/p/43041436
    #http://www.360doc.com/content/17/0321/10/10408243_638692495.shtml(更详细)
    #因为我们的inputs是[batch,steps,inputs]，所以time_major=False，返回结果是
    outputs,final_state = tf.nn.dynamic_rnn(lstm_cell,inputs,dtype=tf.float32) #参数1：LSTM、GRU等的记忆单元。参数2：输入的训练或测试数据
    print("-------------------");
    print("outputs-type:",np.shape(outputs))
    print(outputs)
    print("final_state-tuple:",np.shape(final_state))
    print(final_state)
    print(final_state[0])
    print(final_state[1])  #返回的是ht，代表了outputs中的每一个序列中，序列长度最后一个cell的输出结果全部组合在一起，成为了ht
    #inputs=(?,28,28)  [batch,steps,inputs]
    #lstm_size=(100)隐藏层单元数量
    #outputs=(?,28,100) 相当于权值向量（输入层到隐藏层）
    #final_state=(2,?,100)其中2代表ct和ht 
    print("+++++++++++++++++++");
    results = tf.nn.softmax(tf.matmul(final_state[1],weights)+biases)
    return results
    
    
#计算RNN的返回结果
prediction = RNN(x,weights,biases)
#计算损失函数
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y,logits=prediction))
#使用Adam进行优化
train_Step = tf.train.AdamOptimizer(1e-4).minimize(cross_entropy)
#求准确率
accuracy = tf.reduce_mean(tf.cast(tf.equal(tf.argmax(y,1),tf.argmax(prediction,1)),tf.float32))

#初始化
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for iter in range(6):
        for batch in range(n_batch):
            batch_x,batch_y = mnist.train.next_batch(batch_size)
            sess.run(train_Step,feed_dict={x:batch_x,y:batch_y})
            
        acc = sess.run(accuracy,feed_dict={x:mnist.test.images,y:mnist.test.labels})
        print("Iter:%d Testing Accuracy:%f"%(iter,acc))
```
